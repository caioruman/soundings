```python
from datetime import datetime, timedelta
import numpy as np
import sys
import time
import glob
```

```python
stations = open('stations_arctic_step4.txt', 'r')
# Loop throught all the stations
f1 = open('latlonlist_v2.txt', 'w')
for line in stations:
    aa = line.split(';')
    stn_number = aa[0].replace("''", '')

    if stn_number == "#":
        #skip line
        print("skipped ",aa[2], stn_number)
        continue
    print(aa)
    stn_name = aa[1].replace('(', '').replace(')', '').replace(' ', '_').replace("''", '').replace("\n", '').replace("/", '_')

    #print(stn_name, stn_number)
    
    #Open txt file
    #arq = 
    arquivos = glob.glob("{0}/{1}.*.txt".format(stn_name, stn_number))
    print()
    
    f = open(arquivos[-1], 'r')
    
    s1 = 'latitude'
    s2 = 'longitude'    
    for line in f:
        if s1 in line:
            print(line)
            l = line.strip().replace(':', '').split(' ')
            lat_1 = l[1]
            lat_2 = l[2]
            f1.write(f"{stn_number};{stn_name};{lat_1};{lat_2};")            
        if s2 in line:
            print(line)
            l = line.strip().replace(':', '').split(' ')
            lat_1 = l[1]
            lat_2 = l[2]
            f1.write(f"{lat_1};{lat_2}\n")            
            break    
    f.close()
f1.close()    
```
